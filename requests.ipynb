```python
#Load env variables
from dotenv import load_dotenv
load_dotenv()
```

```python
#Create an API Client
from anthropic import Anthropic
client = Anthropic()
model= "claude-sonnet-4-0"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages):
    response = client.messages.create(
        model=model,
        max_tokens=1000,  # correct parameter name
        messages=messages
    )
    return response.content[0].text
```

```python
#Make a starting list of messages
messages = []
# Add in the initial user question of paradoxes
add_user_message(messages, "define paradox in one sentence")
#Pass the list of messages into 'chat' to get an answer
answer = chat(messages)
#Take the answer and add it to the list of messages as an assistant message
add_assistant_message(messages, answer)
messages
#Add in the follow up question from the user
add_user_message(messages, "give me an example")
#Call chatagain with the list of messages to get a final answer
answer= chat(messages)
answer
```
