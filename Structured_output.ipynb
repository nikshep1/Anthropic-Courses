````python
#Load env variables
from dotenv import load_dotenv
load_dotenv()
````

````python
#Create an API Client
from anthropic import Anthropic
client = Anthropic()
model= "claude-sonnet-4-0"
````

````python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages, system =None, temperature=0.1, stop_sequences=[]):
    params = {
        "model": model,
        "max_tokens": 1000,  
        "messages": messages,
        "temperature": temperature,
        "stop_sequences": stop_sequences
    }
    if system:
        params["system"] = system

    response = client.messages.create(**params)
    return response.content[0].text
````

````python
#message prefilling to direct claude response 
messages = []

add_user_message(messages, "Is tea or coffee better at breakfast?")
add_assistant_message(messages, "Coffee is better because")

answer = chat(messages)

answer
````

````python
#stop sequences - forcing claude to stop generating a response immediately when it generates a specific string of characters
messages = []

add_user_message(messages, "Count from 1 to 10")

answer = chat(messages, stop_sequences=[", 5"])

answer
````

````python
#using both to make sure claude only provides the response we want - no extra commentary, headers and styling
messages = []

add_user_message(messages, "Generate a very short Event Bridge rule in json format")
add_assistant_message(messages, "```json")

answer = chat(messages, stop_sequences=["```"])

answer
````
