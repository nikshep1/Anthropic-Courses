```python
#Load env variables
from dotenv import load_dotenv
load_dotenv()
```

```python
#Create an API Client
from anthropic import Anthropic
client = Anthropic()
model= "claude-sonnet-4-0"
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)


def chat(messages):
    response = client.messages.create(
        model=model,
        max_tokens=1000,  # correct parameter name
        messages=messages
    )
    return response.content[0].text
```

```python
# Make an initial list of messages
messages = []
#Use a 'while True' loop to run the chatbot forever
while True:
    #Get user input
    user_input = input("You: ")
    #Add the user input to the messages list
    add_user_message(messages, user_input)
    #Call Claude with the 'chat' function
    answer = chat(messages)
    #Add the assistant message to the messages list
    add_assistant_message(messages, answer)
    #Print the assistant's response
    print ("-------")
    print("Assistant:", answer)
    print ("-------")

```
